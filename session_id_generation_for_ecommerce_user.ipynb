```python
session=dict()
count=0
for i in range(len(data)):
    if (data.iloc[i,1]) in session.keys():
        if data.iloc[i,0]-session[data.iloc[i,1]][0]>=900:
            data.iloc[i,6]=count
            session[data.iloc[i,1]][1]=count
            count+=1
        else:
            data.iloc[i,6]=session[data.iloc[i,1]][1]
            session[data.iloc[i,1]][0]=data.iloc[i,0]
    else:
        count+=1
        data.iloc[i,6]=count
        session.update({data.iloc[i,1]:[data.iloc[i,0],count] })
    if( i % 500 ==0 ):
        print( f'at {i} th ieration with count {count}')
# session
```
